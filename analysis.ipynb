```python
import pandas as pd
import matplotlib.pyplot as plt
import io
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
print(uploaded.keys())
```

```python
df = pd.read_csv(io.BytesIO(uploaded['top_15_subcategories (3).csv']))
print(df.head())
```

```python
plt.figure(figsize=(12, 6))
plt.bar(df['Sub-Category'], df['total_sales'], color='skyblue')
plt.title('Выручка по подкатегориям (топ-15)')
plt.xlabel('Подкатегория')
plt.ylabel('Выручка')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
profit_counts = df['profit_category'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(profit_counts, labels=profit_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Распределение подкатегорий по рентабельности')
plt.show()
```

```python
df.to_csv('top_15_subcategories.csv', index=False)
```
